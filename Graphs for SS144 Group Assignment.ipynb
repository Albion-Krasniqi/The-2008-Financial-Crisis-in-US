```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import matplotlib.dates as mdates
```

```python
us_data = pd.read_csv('us_data.csv')
us_data.head()
```

```python
us_data.date = pd.to_datetime(us_data.date)

fig = plt.figure(figsize=(7,4))

plt.plot(us_data['date'], us_data['gdp_growth'], label="GDP Annual Growth Rate")

plt.grid(alpha=0.25)

# As we are plotting annual variability, put years as the y ticks
plt.gca().xaxis.set_major_locator(mdates.YearLocator()) 
# And format the ticks to just show the year
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

plt.title("GDP Annual Growth Rate, 2004-2012")
plt.xlabel("Years")
plt.ylabel("Percentage (%)")
plt.legend()
plt.show()
```
